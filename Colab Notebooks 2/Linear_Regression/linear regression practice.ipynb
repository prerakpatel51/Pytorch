```python
import torch
from torch.nn import Linear
```

```python
w=torch.tensor(3.0,requires_grad=True)

b=torch.tensor(1.0,requires_grad=True)
```

```python
def forward(x):
  y= w * x + b
  return y
```

```python
# x=torch.tensor(2)
# forward(x)
# can calculate for multiple x values
x=torch.tensor([[2],[3]])
forward(x)
```

```python
torch.manual_seed(1)
model=Linear(in_features=1,out_features=1)
print(model.weight,model.bias)
```

```python
x=torch.tensor([2.0])
x=torch.tensor([[2.0],[4.0]])

print(model(x))
```
