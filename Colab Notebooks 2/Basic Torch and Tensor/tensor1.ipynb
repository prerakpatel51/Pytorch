```python
#  derivation like slope of the function at x=2 can be found in this way
x=torch.tensor(2.0,requires_grad=True)
y= 9*x**4 + 2*x**3+3*x**2+6*x+1
y.backward()
x.grad
```

```python
# give y=x**2 +z**3 let partially differentiate it using tensor
# to find f'(x,z)=f'(1,2) =(2x)(with respect to x) +(3z**2)(with respect to z)
x=torch.tensor(1.0,requires_grad=True)
z=torch.tensor(2.0,requires_grad=True)

y=x**2 +z**3
y.backward()
print(x.grad)
print(z.grad)
```
